```python
df = pd.read_pickle('ethereum_tweets.pkl') #reading the pickle file
```

# Clean tweets

```python
# function to clean all text
def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    # if there's more than 1 whitespace, then make it just 1
    text = re.sub('\s+', ' ', text)
    # removing any usernames
    text = re.sub('(@[^\s]+)', '', text)
    # getting rid of links
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)

    return text

round1 = lambda x: clean_text_round1(x)
```

```python
# appling data cleaning function to `tweet` column
df['clean_tweets'] = df['text'].apply(round1)
```

```python
collist = ['time', 'text', 'created_at', 'clean_tweets']
```

```python
df[collist].head(30)
```

# Apply VADER sentiment on cleaned tweets

```python
analyser = SentimentIntensityAnalyzer()
```

```python
# function to calculate polarity scores
pol = lambda x: analyser.polarity_scores(x)
```

```python
# creating new column 'polarity' in clean_df
df['polarity'] = df['clean_tweets'].apply(pol)
```

```python
df.head()
```

```python
df2 = pd.concat([df.drop(['polarity'], axis=1), df['polarity'].apply(pd.Series)], axis=1) #extract polarity scores into different columns
```
